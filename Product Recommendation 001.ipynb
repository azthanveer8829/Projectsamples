```python
# I will prepare recommendataions rules for United kingdom retail data.
# Seperating transactions for United Kingdom

data[data['Country']=='United Kingdom']['Country'].count()
```

```python
data[data['Country']=='United Kingdom']
.groupby(['InvoiceNo', 'Description'])['Quantity']
```

```python
mybasket = data[data['Country']=='United Kingdom'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
mybasket.head()
```

```python
# converting all positive values to 1 and everything else to 0

def my_encode_units(x):
    if x <=0:
        return 0
    if x >= 1:
        return 1

my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.head(3)
        
```

```python
my_basket_sets.columns
```

```python
my_basket_sets.drop(['amazon', 'check', 'damages'], inplace=True, axis=1)
```

```python
my_basket_sets.columns
```

## Training Model 

```python
# Generating frequent item sets

my_frequent_itemsets = apriori(my_basket_sets, min_support=0.05, use_colnames=True)
```

```python
my_frequent_itemsets.head(15)
```

```python
# Genarating rules

my_rules = association_rules(my_frequent_itemsets, metric='lift', min_threshold=1)
```

```python
# viewing top rules
my_rules
```

## Making recommendations

```python
my_basket_sets['HAND WARMER BIRD DESIGN'].sum()
```

```python
# filtering rules based on condition

my_rules[ (my_rules['lift'] >= 6) & (my_rules['confidence'] >= 0.5) ]
```
